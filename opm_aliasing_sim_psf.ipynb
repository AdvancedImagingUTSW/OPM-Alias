```python
import numpy as np
from numpy.fft import fft2, ifft2, fftshift, ifftshift, fftfreq
import matplotlib.pyplot as plt


from skimage.transform import resize
```

```python
def gauss2Drot(n, m, s, theta):
    
    x, y = np.meshgrid(range(n[0]), range(n[1]))

    x = x - x.mean()
    y = y - y.mean()

    coords = np.vstack((
        x.flatten(),
        y.flatten(),
        np.zeros(n[0] * n[1])
    ))

    t = np.radians(theta)
    rotz = np.array([
        [np.cos(t), -np.sin(t), 0],
        [np.sin(t), np.cos(t), 0],
        [0, 0, 1],
    ])

    coords = np.matmul(rotz, coords)
    x = np.reshape(coords[0], [n[1],n[0]])
    y = np.reshape(coords[1], [n[1],n[0]])

    sx, sy = s
    mx, my = m

    ex = (x - mx)/sx
    ey = (y - my)/sy

    e = ex*ex + ey*ey
    return np.exp(-e)

def fftconv(x, y):
    a = fft2(x)
    b = fft2(y)
    
    z = np.real(ifftshift(ifft2(a * b)))

    return z

def pwr(x):
    p = np.abs(x)
    p = np.log10(p + 1e-20)
    return fftshift(p)

def circ(im, r=10, t=2):
    
    x, y = np.meshgrid(range(im.shape[1]), range(im.shape[0]))
    x = x - x.mean()
    y = y - y.mean()

    c = (x**2 + y**2 < (r+t)**2).astype(int)
    c -= (x**2 + y**2 < r**2).astype(int)   

    return c 

def wiener_filt(x, alpha=0.005):
    X = fft2(x)
    X = X / (X**2 + alpha)
    return X

def rms(a : np.ndarray, b : np.ndarray):
    a = a / a.sum()
    b = b / b.sum()
    
    err = a - b
    err = err**2

    return np.sqrt(err.mean())

def ctrcrop(x, w=48, h=48):
    cy, cx = x.shape
    cx = int(cx/2)
    cy = int(cy/2)

    return x[(cy-h):(cy+h), (cx-w):(cx+w)]

def norm(x):
    x -= x.min()
    return x / x.max()

from matplotlib.axes import Axes
def ksp_ticks(ax : Axes ):
    ax.apply_aspect()

    xt = ax.get_xticks()
    yt = ax.get_yticks()

    nx = len(xt)
    ny = len(yt)
    
    dx = xt[1] - xt[0]
    dy = yt[1] - yt[0]

    ax.set_xticklabels([f"{x:.4f}" for x in fftshift(fftfreq(nx, d=dx))])
    ax.set_yticklabels([f"{y:.3f}" for y in fftshift(fftfreq(ny, d=dy))])
```

```python
# ---- Parameters ----

# image dimensions
m = 65
n = 65

# psf
pSize = [12,2]
tiltAngle = -45

# downsampling
dsFactor = 4
```

```python
# ---- SIMULATION ----

psf = gauss2Drot([m,n], [1,1], pSize, tiltAngle)
psf = psf / psf.sum()

im = psf

# domains:
ext = [-m/2, m/2, -n/2, n/2]

fig_1, ax = plt.subplots(2,4, figsize=(24,12))

ax[0][0].imshow(im, extent=ext)
ax[0][0].set_xlabel("x")
ax[0][0].set_ylabel("y")
_ = ax[0][0].set_title(f"PSF: {pSize[0]} x {pSize[1]}, {tiltAngle}-deg", fontsize=20)

# applying comb function
comb = np.zeros_like(im)
comb[::dsFactor] = 1
im_comb = im * comb

im_comb = im_comb / im_comb.sum()

# downsampled image, simulating undersampled data collection
im_ds = im_comb[im_comb.any(1) > 0]

# plot downsampled image and fft
ax[0][1].imshow(im_ds, extent=[-m/2, m/2, -n/dsFactor/2, n/dsFactor/2])
_ = ax[0][1].set_title(f"undersampled PSF ({dsFactor}X)", fontsize=20)

ims = ax[1][1].imshow(pwr(fft2(im_ds)), extent=[-m/2, m/2, -n/dsFactor/2, n/dsFactor/2])
_ = ax[1][1].set_title(f"FT[undersampled]", fontsize=20)

ksp_ticks(ax[1][1])

# plot comb
ax[0][2].imshow(im_comb)
_ = ax[0][2].set_title(f"im_comb: \"comb\" upsampling", fontsize=20)

# G
G = fft2(im_comb)

sy, sx = G.shape
alpha = np.radians(90-tiltAngle)

x, y = np.meshgrid(np.arange(sx), np.arange(sy))
x_mid = x.mean()
y_mid = y.mean()
x = x.astype(float) - x_mid
y = y.astype(float) - y_mid

# shift for even dsFactor case
y_mid -= float(not np.mod(dsFactor, 2)) * sy/dsFactor/2

xyAspect = sx/sy

ax[1][0].imshow(pwr(fft2(im)), extent=ext)
ax[1][0].set_xlabel("Kx")
ax[1][0].set_ylabel("Ky")
_ = ax[1][0].set_title("FT[PSF]", fontsize=20)

ksp_ticks(ax[1][0])

ax[1][2].imshow(pwr(G), extent=ext)
y_i = np.arange(G.shape[0])
l_y = y_i[np.mod(y_i, int(sy/dsFactor)) == 0]
ax[1][2].hlines(l_y - y_mid, ext[0], ext[1], linestyles='--', color='w')

ax[1][2].plot(x[0], np.tan(alpha) * x[0][::-1]/xyAspect + sy/dsFactor/2, color='w')
ax[1][2].plot(x[0], np.tan(alpha) * x[0][::-1]/xyAspect - sy/dsFactor/2, color='w')
ax[1][2].plot(x[0], np.tan(alpha) * x[0][::-1]/xyAspect, '--', color='w')

ax[1][2].set_xlim((ext[0], ext[1]))
ax[1][2].set_ylim((ext[2], ext[3]))
_ = ax[1][2].set_title(f"G = FT[im_comb]", fontsize=20)

ksp_ticks(ax[1][2])

# pick out central lobe:
mask = y > np.tan(alpha) * x/xyAspect + sy/dsFactor/2
mask += y < np.tan(alpha) * x/xyAspect - sy/dsFactor/2
mask = np.logical_not(mask)

G_mask = ifftshift(mask * fftshift(G))

ax[1][3].imshow(pwr(G_mask), extent=ext)
_ = ax[1][3].set_title("Mask x G", fontsize=20)

ksp_ticks(ax[1][3])

# IFT
g_recon = np.real(ifft2(G_mask))

ax[0][3].imshow(g_recon, extent=ext)
_ = ax[0][3].set_title("g_recon", fontsize=20)
```

```python
fig_2, ax = plt.subplots(1,4, figsize=(24,6))

ax[0].imshow(im, extent=ext)
ax[0].set_title(f"ground truth PSF", fontsize=20)
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

# plot downsampled image and fft
ax[1].imshow(im_ds, extent=[-m/2, m/2, -n/dsFactor/2, n/dsFactor/2])
ax[1].set_title(f"undersampled PSF ({dsFactor}X)", fontsize=20)

# interpolated
g_interp = resize(im_ds, im.shape, order=3)

ax[2].imshow(g_interp, extent=ext)
ax[2].set_title("cubic spline interpolation", fontsize=20)

p = ax[2].get_position().bounds

ax_error = fig_2.add_axes([p[0] + 0.6*p[2], p[1] + 0.52*p[3], 0.36*p[2], 0.36*p[3]])
abs_err = np.abs(norm(im) - norm(g_interp))
err_im = ax_error.imshow(abs_err)

ax_error.set_xticks([])
ax_error.set_yticks([])
_ = [s.set_color('w') for s in ax_error.spines.values()]

cbar = plt.colorbar(err_im, location='top', fraction=0.05, pad=0.04)
cbar.ax.tick_params(colors='w')
plt.title("|psf - interp|", color='w', x=0.5, y=-0.2)

# reconstructed
g_recon = np.real(ifft2(G_mask))

ax[3].imshow(g_recon, extent=ext)
ax[3].set_title("comb-and-mask recon", fontsize=20)

p = ax[3].get_position().bounds

ax_error = fig_2.add_axes([p[0] + 0.6*p[2], p[1] + 0.52*p[3], 0.36*p[2], 0.36*p[3]])
abs_err = np.abs(norm(im) - norm(g_recon))
err_im = ax_error.imshow(abs_err)

ax_error.set_xticks([])
ax_error.set_yticks([])
_ = [s.set_color('w') for s in ax_error.spines.values()]

cbar = plt.colorbar(err_im, location='top', fraction=0.05, pad=0.04)
cbar.ax.tick_params(colors='w')
plt.title("|psf - recon|", color='w', x=0.5, y=-0.2)
```
